```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv('iris.csv')
iris
```

```python
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('sepal length')
plt.ylabel('petal length')
```

```python
#control the figure size
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='summer',alpha=.7)
plt.xlabel('sepal length')
plt.ylabel('petal length')
plt.colorbar()
```

```python
plt.scatter(df['avg'],df['strike_rate'])
for i in range(df.shape[0]):
 plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
```

```python
plt.figure(figsize=(18,10))
plt.scatter(df['avg'],df['strike_rate'],s=df['runs'])
for i in range(df.shape[0]):
 plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
```

```python
#horizontal and vertical line
plt.figure(figsize=(18,10))
plt.scatter(df['avg'],df['strike_rate'],s=df['runs'])
plt.axhline(130,color='red')
plt.axvline(30,color='red')
for i in range(df.shape[0]):
 plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
#contour plot->it give 3_d to 2-D graph
z=xx**2+yy**2
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
#contourf
z=xx**2+yy**2
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
