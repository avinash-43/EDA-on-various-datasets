```python
#mutliple bar chart
df=pd.read_csv('batsman_season_record.csv')
df.head()
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
plt.bar(np.arange(df.shape[0])-.2,df['2015'],width=.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=.2)
plt.bar(np.arange(df.shape[0])+.2,df['2017'],width=.2)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
#bar chart 
child=[20,30,45,35]
colors=['redredredred','greengreengreen','yellowyellowyellow','pink']
plt.bar(colors,child,color='black')
```

```python
child=[20,30,45,35]
colors=['redredredred','greengreengreen','yellowyellowyellow','pink']
plt.bar(colors,child,color='black')
plt.xticks(rotation='vertical')
```

```python
#stacked bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2016']+df['2015'],label='2017')
plt.legend()
plt.show()
```
